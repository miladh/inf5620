## Mathematical problem
Consider a vibration problem

$$
u''(t) + \omega^2 u(t) = 0,\quad u(0)=I,\ u'(0)=V
$$

Here, $\omega$, $I$, and $V$ are given data. We discretize the ODE by centered diferences,

$$
[D_tD_t u + \omega^2 u =0 ]^n
$$



---
##Part I

*Explain how the stability and the accuracy of this scheme can be investigated via exact solutions of the discrete equations, and quote the main results. Illustrate the numerical problems that can arise from this scheme.*

**Numerical problems: **

   * Phase error which is reduced by reducing the time step. Total error grows with time (accuracy).
    
   * An unstable growing amplitude that can occur if $\Delta t$ is too large (stability).


**Exact numerical solution: **
     
   * $u^n=A^n = Ie^{i\tilde\omega \Delta t n} = Ie^{i\tilde\omega t}=  I\cos (\tilde\omega t) + iI\sin(\tilde \omega t)$

**Numerical frequency: ** 

   * $\tilde\omega = \pm \frac{2}{\Delta t}\sin^{-1}\left(\frac{\omega\Delta t}{2}\right) =\pm \omega+ {\mathcal{O}(\Delta t^2)}$
    
   * $\tilde\omega \neq \omega$ always and  $\tilde\omega > \omega$.
    
   * Total phase error due to the frequency error grows linearly with time:
       * Define the phase error as time lag between the peak $I$ in the exact solution and the corresponding peak in the approximation after $m$ periods of oscillations. Phase error is linear in $m$: 
           * $t_m$: time when $u_e$ reach peak norm.
           * Location: $2\pi m \quad\Rightarrow\quad \omega t_m = 2 \pi m \quad\Rightarrow\quad t_m = \frac{2\pi}{\omega}m $
           * $t_m - \tilde t_m = 2 \pi m(\omega^{-1}- \tilde \omega^{-1})$
    
**Discretization parameter: **     
    
   * $p=\omega\Delta t$, number of time steps per period of the oscillations. 
       * $P=N_P\Delta t=2\pi/\omega$, where $N_p$ is the number of time steps during a period.
       * $\omega \Delta t = 2\pi/N_p$
        
**Stability: **  

   * Constant amplitude demands $\tilde \omega$ to be real
   * Can be complex when $\sin^{-1}(x), \quad |x| > 1$ 
        
       * $\tilde\omega = \tilde\omega_r +i\tilde\omega_i$(negative imaginary part for $x>1$)
        
       * Exponential growth in time: $\exp{(-\tilde\omega_i t)}$ with $\tilde\omega_i < 0$

       * $\frac{\omega\Delta t}{2} \leq 1\quad\Rightarrow\quad \Delta t \leq \frac{2}{\omega}$
    
    * Would it be meaningful to take larger time steps to speed up computations? 
    
       * At the stability limit: 
           * $\tilde\omega = \frac{2}{\Delta t}\sin^{-1}(\omega\Delta t/2) = \frac{2}{\Delta t}\sin^{-1} 1 = \pi/\Delta t$
           * $\tilde P=2\pi/\tilde\omega = 2\Delta t$ 
            
            * Just one time step $\Delta t$ between a peak and a through in the numerical solution. 
            * This is the shortest possible wave that can be represented in the mesh. In other words, it is not meaningful to use a larger time step than the stability limit.

##Part II

*We now consider a 1D wave equation $u_{tt} = c^2u_{xx}\rightarrow [D_tD_t u = c^2 D_xD_x]^n_i$ with some appropriate boundary and initial conditions. Explain how the stability and accuracy of a centered difference scheme can be investigated via exact solutions of the discrete equations. Quote the main results. *


**Exact numerical solution: **
     
   * $u^n_q=e^{i(kx_q - \tilde\omega t_n)}$
     
   * $\tilde \omega \neq \omega = kc $
     
   * $\sin\left(\frac{\tilde\omega\Delta t}{2}\right) = C\sin\left(\frac{k\Delta x}{2}\right),$

**Discretization parameter: **     
    
   * $p=k\Delta x/2$
       * How many spatial mesh points we have per wave length in space of the wave component with frequency $k$.
       * Reflects how well the spatial variation of the wave component is resolved in the mesh.
       * Wave components with $\lambda = 2\pi/k < 2\Delta x$ so $p>\pi/2$ is non sense.

**Stability: **  

   * $\sin\left(\frac{\tilde\omega\Delta t}{2}\right) = C\sin\left(\frac{k\Delta x}{2}\right)$
   
   * $\frac{k\Delta x}{2} = m\frac{\pi}{2},\quad m\in\mathbb{Z}$
    
   * Shortest wave length: $m=1 \rightarrow k\Delta x=\pi \quad \Rightarrow \lambda = 2\pi/k = 2\Delta x$
       * $C\leq 1$ for shortest wave

**Numerical dispersion relation: ** 
    
   * $\tilde\omega = \frac{2}{\Delta t}\sin^{-1}\left( C\sin\left(\frac{k\Delta x}{2}\right)\right)$
   * $C = \frac{c\Delta t}{\Delta x} = 1 \quad \Rightarrow \quad \tilde\omega = \omega$

##Part IV

*Explain how a truncation error analysis is carried out for the problem in a). Find correction terms such that the order of the scheme becomes $\Delta t^4$.*

**Idea: **

   * To what extent the exact solution $u_e$ fits the discrete equations? 
   * $u_e$ is in general not a solution of $\mathcal{L}_\Delta(u)=0$
   * Residual $\mathcal{L}_\Delta(u_e)=R$
   * A small $R$ means intuitively that the discrete equations are close to the differential equation, and then we are tempted to think that $u_n$ must also be close to $u_e(t_n)$.
    
   * Reveals the accuracy of the various building blocks in the numerical method and how each building block impacts the overall accuracy. 
  
  
**Residual: ** 
   
  * $\lbrack D_tD_t u \rbrack^n = \frac{u^{n+1} - 2u^{n} + u^{n-1}}{\Delta t^2} = u''(t_n) + R^n$
  * $[D_tD_t {u_{\small\mbox{e}}} + \omega^2{u_{\small\mbox{e}}} =R]^n$
    
   * $u(t_{n+1}) = u(t_n) + u'(t_n)\Delta t + {\frac{1}{2}}u''(t_n)\Delta t^2 +{\frac{1}{6}}u'''(t_n)\Delta t^3 + {\frac{1}{24}}u''''(t_n)\Delta t^4 +  {\frac{1}{120}}u'''''(t_n)\Delta t^5 +{\mathcal{O}(\Delta t^6)}$
    
   * $u(t_{n-1}) = u(t_n) - u'(t_n)\Delta t + {\frac{1}{2}}u''(t_n)\Delta t^2 -{\frac{1}{6}}u'''(t_n)\Delta t^3 + {\frac{1}{24}}u''''(t_n)\Delta t^4 -  {\frac{1}{120}}u'''''(t_n)\Delta t^5+  {\mathcal{O}(\Delta t^6)}$
    
   * $R^n =  \frac{1}{12}{u_{\small\mbox{e}}}''''(t_n)\Delta t^2 + {\mathcal{O}(\Delta t^4)}$
   
   
**Initial conditions: ** 

   * $u'(0)=V$ involves a centered difference with an ${\mathcal{O}(\Delta t^2)}$.
   * $R^n = \frac{1}{6}u'''(0)\Delta t^2 + {\mathcal{O}(\Delta t^3)}$
   * Investigate the impact of approximating $u'(0)$ on overall convergence rate of the scheme:
       * $[D_{2t}u=V]^0$ gives $\Delta t^2$ 
       * $[D_t^+u=V]^0$ gives $\Delta t$ 


**Correction term: ** 

   * $[D_tD_t {u_{\small\mbox{e}}} + \omega^2{u_{\small\mbox{e}}} =C + R]^n$
   * $C^n$ must be chosen to cancel $\Delta t^2$ in $R^n \quad \Rightarrow \quad C^n = \frac{1}{12}{u_{\small\mbox{e}}}''''(t_n)\Delta t^2{\thinspace .}$ 
    
   * Using $u'''' = \omega^4 u$ we obtain $[D_tD_t u + \omega^2(1 - \frac{1}{12}\omega^2\Delta t^2)u=0]^n$

##Part V

*Explain how a truncation error analysis is carried out for the problem in b). Find correction terms such that the order of the scheme becomes $\Delta t^4$,$\Delta x^4$.*

**Residual: ** 
   
   * $[D_t D_t {u_{\small\mbox{e}}} = c^2 D_xD_x {u_{\small\mbox{e}}} + R]^n_i$
   * $ [D_t D_t{u_{\small\mbox{e}}}]_i^n = {u_{\small\mbox{e}, tt}}(x_i,t_n) +\frac{1}{12}{u_{\small\mbox{e}, tttt}}(x_i,t_n)\Delta t^2 + {\mathcal{O}(\Delta t^4)},$
   * $[D_xD_x{u_{\small\mbox{e}}}]_i^n = {u_{\small\mbox{e}, xx}}(x_i,t_n) +\frac{1}{12}{u_{\small\mbox{e}, xxxx}}(x_i,t_n)\Delta x^2 + {\mathcal{O}(\Delta x^4)}$
   * $ R^n_i = \frac{1}{12}{u_{\small\mbox{e}, tttt}}(x_i,t_n)\Delta t^2 -
     c^2\frac{1}{12}{u_{\small\mbox{e}, xxxx}}(x_i,t_n)\Delta x^2 +
     {\mathcal{O}(\Delta t^4,\Delta x^4)}$


**Correction term: ** 
    
   * $[D_t D_t {u_{\small\mbox{e}}} = c^2 D_xD_x {u_{\small\mbox{e}}} + C + R]^n_i$
       * $C^n$ must be chosen to cancel $\Delta t^2, \Delta x^2$ in $R^n \quad \Rightarrow \quad C = \frac{1}{12}\left( u_{tttt}\Delta t^2 - c^2u_{xxxx}\Delta x^2\right),$ 
    
       * Using $u_{tttt} = c^2u_{xxtt},\quad u_{xxxx} = c^{-2}u_{ttxx} \quad \Rightarrow \quad$ $C^n_i = \frac{1}{12}((c^2\Delta t^2 - \Delta x^2) [D_xD_xD_tD_t u]^n_i$
    * Leads to coupled equation, tridiagonal system should be solved.
    * This is in principle straightforward, but it results in an implicit finite difference schemes, while we had a convenient explicit scheme without the correction terms.
